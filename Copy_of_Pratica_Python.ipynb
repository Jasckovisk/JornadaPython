#1º Exercício Atividade Prática

```python
def calculadora_desconto(quantidade, unidade): # dentro dos parenteses em azul são os argumentos da função.
    if quantidade <= 9:
        desconto = 0
    elif quantidade <= 99:
        desconto = 5
    elif quantidade <= 999:
        desconto = 10
    else:
        desconto = 15
    valor = unidade * quantidade
    desconto_valor = valor * (desconto / 100)
    valor_desconto = valor - desconto_valor
    return valor, valor_desconto, desconto

print("Bem-vindos à loja do Jackson Douglas")
unidade = float(input("Valor do Produto: "))
quantidade = int(input("Quantidade do Produto: "))
valor, valor_desconto, desconto = calculadora_desconto(quantidade, unidade) # Essa linha é para chamar a função e atribuir com o sinal de '=' as váriaveis valor/valor_desconto e desconto!

print("O valor sem desconto foi de: R${:.2f}".format(valor))
print("O valor com desconto foi de: R${:.2f} (Desconto {}%)".format(valor_desconto, desconto)) # Dificuldade encontrada! Dar atenção ao conteúdo de formatação confira o PDF se precisar"
```

#3º Exercício Atividade Prática

```python
#1º Função para dimensionar o objeto
def dimensoes_objt(): # definindo as entradas do problema.
  while True:
    try:
      altura = float(input('Digite a altura do objeto em cm: '))
      comprimento = float(input('Digite o comprimento do objeto em cm: '))
      largura = float(input('Digite a largura do objeto em cm: '))

      volume = altura * comprimento * largura
      print('O volume é do objeto é', volume ,'cm³')

      if volume < 1000:
        return 10
      elif volume < 10000:
        return 20
      elif volume < 30000:
        return 30
      elif volume < 100000:
        return 50
      else:
        print('Desculpe não transportamos objetos tão grandes.')
        print('Entre com as dimensões novamente.')
    except ValueError:
        print('Você digitou alguma dimensão do objeto com valor não numérico')
        print('Por favor entre com as dimenções do objeto novamente')
#2º Função para definir o peso do objeto
def peso_objt():
  while True:  # Começo do laço
    try: # tentar
      peso = float(input('Digite o peso do objeto em kg: '))

      if peso <= 0.1: # se for menor ou igual a 0.1
        return 1 # retornara o valor 1
      elif peso <= 1:
        return 1.5
      elif peso <= 10:
        return 2                       
      elif peso <= 30:
        return 3
      else: # caso o peso seja diferente das opções acima aparece a mensagem de erro!
        print('Muito peso! Vai da não chefe!')
        print('Digite o peso novamente.')
    except ValueError: # exceto para o erro de ValueError, quando ocorrer esse erro mostrara o print abaixo do except!
        print('Você digitou peso com valor não numérico. Digite o peso novamente.')  
#3º Função para definir a rota do objeto
def rota_objt():
  while True: 
    print('Rota do objeto:')
    print('BR - De Brasília para Rio de Janeiro')
    print('BS - De Brasília para São Paulo')
    print('RB - De Rio de Janeiro para Brasília')
    print('SR - De São Paulo para Rio de Janeiro')
    print('SB - De São Paulo para Brasília')
    
    rota = input().upper() # Esse comando transforma o input "rota" em minusculo, para evitar erros caso o user digite em minusculo.
    if rota in ['BR','BS','RB','RS','SR','SB']:
      if rota in ['RS','SR']:
        return 1
      elif rota in ['BS','SB']:
        return 1.2
      else: # qualquer rota que nao for as listadas acima "bs,sb ou rs,sr" vai ser o restante da função no caso o "sb,bs"
        return 1.5
    else: # qualquer rota não for o restante da lista.
      print('Rota inválida!')
      print('Por favor entre com a rota desejada novamente.')

#Programa Principal
print('Olá Bem Vindo a Companhia de logística Jackson Douglas S.A')      
      
#4º Função para calcular o total dando call para as funções anteriores.
def calcular_total():
  valor_dimensoes = dimensoes_objt()
  valor_peso = peso_objt()
  valor_rota = rota_objt()

  total = valor_dimensoes * valor_peso * valor_rota
  print('Valor do envio: R$', f'{total:.2f}', '(Dimensões:',valor_dimensoes,'* peso:',valor_peso,'* rota:',valor_rota,')') # Formatação mais recente após o 3.6 em versões anteriores utilizar o {:.f} .format{exemplo}
  return
  
calcular_total() # Aprendi que posso chamar todas as funções apenas com a ultima pois para rodar a ultima função necessáriamente precisa rodar as anteriores.
```

```python
#1º Função para cadastro da peça
def cadastro_peca(codigo):
  peca = {} #Dicionário vazio
  codigo_str = '{:03}'.format(int(codigo)) # linha para formatar o valor em três digitos.
  print('Código da Peça:',codigo_str)
  peca['nome'] = input('Digite o nome da peça: ')
  peca['fabricante'] = input('Digite o fabricante da peça: ')
  peca['valor'] = '{:.2f}'.format(float(input('Digite o valor da peça: '))) # '{:.2f} formatação para que o ponto flutuante seja exibido com 2 casas decimais.

  return codigo_str, peca
#2º Função para consultar as peças cadastradas
def consultar_peca(pecas):
  
  while True:
    print('Menu de Consulta de peça:') # \n você pode usar escape para quebra de linha
    print('1) Consultar todas as peças')
    print('2) Consultar peças por código')
    print('3) Consultar peças por fabricante')
    print('4) Retornar')

    opcao = input('>>')

    if opcao == '1':
      if len(pecas) == 0:
        print('Nenhuma peça cadastrada.')
      else:
          print('Todas as peças:')
          for peca in pecas:
            # \n você pode usar escape para quebra de linha
            codigo = str(peca[0])
            codigo_formatado = '{:03}'.format(codigo) 
            print('Código:', int(peca[0])) # int aqui para tranformar em numero inteiro passando ele de 3 para 1 casa decimal.
            print('Nome:',peca[1]['nome'])
            print('Fabricante:',peca[1]['fabricante'])
            print('Valor:',peca[1]['valor'])
            print('-------')

    elif opcao =='2':
      codigo = input('Digite o código da peça: ')
      for peca in pecas:
        if str(peca[0]) == codigo.zfill(3): #essa linha contem uma variavel do format para formatar com 3 zeros à esquerda.        
          print('-------')
          print('Código:', str(peca[0]).zfill(3))
          print('Nome:', peca[1]['nome'])
          print('Fabricante:', peca[1]['fabricante'])
          print('Valor:', peca[1]['valor'])
          print('-------')
          break

      else:
        print('Peça não encontrada.')

    elif opcao == '3':
      fabricante = input('Digite o fabricante da peça: ')
      print('Peças do fabricante', fabricante)
      encontrada = False
      for peca in pecas:
        if peca[1]['fabricante'] == fabricante:
          encontrada = True        
          codigo = str(peca[0])
          codigo_formatado = codigo.zfill(3) # formatando o código com uma terceira maneira além das vistas anteriormente, caso tenha duvidas consultar o material em PDF sobre formatação de strings.  
          print('Código:', codigo_formatado)
          print('Nome:', peca[1]['nome'])
          print('Valor:', peca[1]['valor'])
          print('-------')
          if not encontrada:
            print('Nenhuma peça encontrada para o fabricante', fabricante + '.')

    elif opcao == '4':
      break

    else:
        print('Opção inválida. Digite novamente.')

#3º Função para remover peças cadastradas
def remover_peca(pecas):
    codigo = input('Digite o código da peça que deseja remover: ')
    codigo = codigo.zfill(3)
    for i, peca in reversed(list(enumerate(pecas))): # Dificuldade encontrada! Procurar conteúdo extra na web, Método de ensino compátivel com meu método de aprendizagem!   
        if peca[0] == codigo:
            del pecas[i]
            print('Peça removida com sucesso!')
            return

    print('Peça não encontrada.')

pecas = []
contador = 1

print('Bem vindo ao Controle de Estoque da Bicicletaria do Jackson Douglas')
#Laço de repetição
while True:
  print('Menu de Controle de Estoque:')
  print('1) Cadastrar Peça')
  print('2) Consultar Peça')
  print('3) Remover Peça')
  print('4) Sair')
  
  opcao = input('>>')

  if opcao == '1':
    print('Você Selecionou a Opção de Cadastrar Peça')
    codigo, peca = cadastro_peca(contador)
    pecas.append((codigo,peca))
    contador += 1
    print('Peça cadastrada com Sucesso!') #linha adicional achei interessante retornar para o usuario que a operação do cadastro deu boa!
  elif opcao == '2':
    consultar_peca(pecas)
  elif opcao == '3':
        print('Você selecionou a Opção de Remover Peça')
        remover_peca(pecas)      
  elif opcao == '4':
    break
  else:
    print('Opção inválida. Digite novamente') 
```
